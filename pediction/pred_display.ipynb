```python
import pandas as pd

# Load your data
df = pd.read_csv("/Users/lancesanterre/so_predict/data/combined_pred/2025-07-25_pitcher_preds_combined.csv")

# Rename columns for clarity
df = df.rename(columns={
    "unique_name": "Pitcher",
    "rf_reg": "RF Prediction",
    "xgb_reg": "XGB Prediction",
    "average_prediction_reg": "Avg Prediction (Reg)",
    "final_prediction_class": "Final Class",
    "mode_class_class": "Mode Class",
    "avg_prob_class": "Avg Probability",
    "hybrid_class_class": "Hybrid Class"
})

# --- Styling functions ---
def highlight_class(val):
    return 'background-color: lightgreen; color: black;' if val == 1 else 'background-color: salmon; color: black;'

def format_prob(val):
    return f"{val:.2f}"

# --- Build styled DataFrame ---
styled = (
    df.style
    .applymap(highlight_class, subset=["Final Class", "Mode Class", "Hybrid Class"])
    .bar(subset=["RF Prediction", "XGB Prediction", "Avg Prediction (Reg)"], color='#87CEEB')
    .format({
        "RF Prediction": "{:.2f}",
        "XGB Prediction": "{:.2f}",
        "Avg Prediction (Reg)": "{:.2f}",
        "Avg Probability": format_prob
    })
    .set_table_styles([{
        'selector': 'th',
        'props': [('background-color', '#2C3E50'), ('color', 'white'), ('font-weight', 'bold')]
    }])
)

styled
```
